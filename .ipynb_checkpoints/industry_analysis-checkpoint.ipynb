```python
import numpy as np
import pandas as pd
```

```python
ind_df = pd.read_csv('inc_cleaned_data.csv', sep='|')
```

```python
ind_df.head()
```

```python
ind_df.loc[:,('prev_revenue')] = ind_df.apply(lambda x: (100/(x['growth']+100)) * x['revenue'], axis=1).values
```

```python
ind_df.info()
```

```python
ind_df.groupby('industry').sum()[['revenue','prev_revenue']]
```

```python
# industry wise average growth
ind_df.groupby('industry').apply(lambda x: ((x['revenue'].sum()-x['prev_revenue'].sum())/x['revenue'].sum())*100)
```

```python
ind_df['industry'].value_counts()
```

```python
ind_df.groupby('industry').describe()['growth']
```
