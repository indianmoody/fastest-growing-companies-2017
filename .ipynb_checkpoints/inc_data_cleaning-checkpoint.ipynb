```python
import numpy as np
import pandas as pd
```

```python
inc_df = pd.read_csv('inc_data.csv', sep='|')
```

```python
inc_df.head(10)
```

```python
analysis_df = inc_df[['Rank_2017', 'Company_Name', 'Revenue_2016', 'Industry', 'Employees', 'Founded', 'Growth_3_yrs', 'City', 'State']]
```

```python
analysis_df.head(10)
```

```python
analysis_df.columns = ['rank','name','revenue','industry','employees','founded','growth','city','state']
analysis_df.head(2)
```

```python
analysis_df.info()
```

```python
analysis_df['industry'].value_counts()
```

```python
analysis_df.groupby('industry').describe()
```

```python
# check if all 'revenue' values are in M (million)
analysis_df['revenue'].apply(lambda x: x[-1]).value_counts()
```

```python
# check if all 'revenue' values are in $
analysis_df['revenue'].apply(lambda x: x[0]).value_counts()
```

```python
#check if there is space between $ sign and amount for any row
analysis_df['revenue'].apply(lambda x: x[1]).value_counts()
```

```python
# and there sum is equal to total number of rows
analysis_df['revenue'].apply(lambda x: x[1]).value_counts().values.sum()
```

```python
# check if all row values for 'revenue' has space between amount and 'M'
analysis_df['revenue'].apply(lambda x: x[-2]).value_counts()
```

```python
# change 'revenue' column format to numerical values by removing $ and M
analysis_df.loc[:,('revenue')] = analysis_df['revenue'].apply(lambda x: x[1:-2]).values
```

```python
analysis_df.head(10)
```

```python
analysis_df['city'].value_counts()
```

```python
analysis_df['revenue']
```

```python
analysis_df.loc[:,('revenue')] = analysis_df['revenue'].apply(pd.to_numeric)
```

```python
analysis_df['revenue'][:10]
```

```python
analysis_df.groupby('industry').describe()['revenue']
```

```python
analysis_df['growth'].apply(lambda x: x[1]).value_counts()
```

```python
analysis_df.sort_values(by='revenue', ascending=False)
```

```python
analysis_df.loc[:,('growth')] = analysis_df['growth'].apply(lambda x: "".join(x[:-1].split(","))).values
```

```python
analysis_df.loc[:,('growth')] = analysis_df['growth'].apply(pd.to_numeric)
```

```python
analysis_df['growth'].head()
```

```python
analysis_df.head()
```

```python
analysis_df[pd.isnull(analysis_df['founded'])]
```

```python
analysis_df.groupby('industry').describe()[['revenue','growth']]
```

```python
analysis_df.to_csv('inc_cleaned_data.csv', index=False, sep='|')
```
